```python
before = ship.copy()
before['Route_Type'] = 'Before'
```

```python
after = ship.copy()

after['Route_Type'] = 'After'

# Increase distance by 40%
after['Distance_km'] = after['Distance_km'] * 1.4

# Add 12 days delay
after['Travel_Days'] = after['Travel_Days'] + 12
```

```python
final_data = pd.concat([before, after])
```

```python
fuel_price = fuel_data['Diesel_Price'].mean()
```

```python
final_data['Fuel_Cost'] = final_data['Distance_km'] * fuel_price * 0.05
#(0.05 is just a simple assumption for fuel usage)
```

Hypothesis Testing (T-test)
In here i checked Did cost really increase after rerouting
so i Comparied  two populations using splitting data into two samples

Group 1 → Before
Group 2 → After

```python
before_cost = final_data[final_data['Route_Type'] == 'Before']['Fuel_Cost']
after_cost = final_data[final_data['Route_Type'] == 'After']['Fuel_Cost']
```

```python
#SciPy is a Python library useful for solving many mathematical equations and algorithms.

from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(after_cost, before_cost)
print("p-value:", p_value)
```

T-test checks Are these two groups different enough, or just random variation

p-value = probability that difference is random

p-value	        Meaning
< 0.05	        Real difference
≥ 0.05	        Might be random

I got p = 0.0
Means There is a VERY strong real difference

so finally proves Shipping cost significantly increased after rerouting

```python
before_time = final_data[final_data['Route_Type'] == 'Before']['Travel_Days']
after_time = final_data[final_data['Route_Type'] == 'After']['Travel_Days']

t_stat, p_value = ttest_ind(after_time, before_time)
print("Time p-value:", p_value)
```

Time p-value = 0.0

Meaning Travel time also significantly increased

used Correlation to Checks Do two variables move together

In here Distance ↑ → Cost ↑

Strong positive relationship

Using Linear Regression
with Statsmodels

```python
#The StatsModels library in Python is a tool for statistical modeling, hypothesis testing and data analysis.
#It provides built-in functions for fitting different types of statistical models, performing hypothesis tests and exploring datasets.

import statsmodels.api as sm

X = final_data['Distance_km']
y = final_data['Fuel_Cost']

# add constant
#It builds equation:
#Fuel_Cost = a × Distance_km + b

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())
```

Distance_km coef = 14.26
Meaning Every 1 km increase adds ~14.26 to cost

R² = 1.0
Meaning Distance explains 100% of cost variation

p-value = 0.0 Relationship is statistically significant”

T-test
Is cost/time increase real?  ✔ YES

Correlation
Do distance and cost move together? ✔ YES

Regression
How much does cost increase with distance? ✔ 14.26 per km


I used T-tests to confirm that differences between scenarios were statistically significant, correlation to examine relationships between variables, and regression analysis to quantify the impact of distance on cost.”

```python
# matplotlib  library use  convert data into  charts

import matplotlib.pyplot as plt

plt.scatter(final_data['Distance_km'], final_data['Fuel_Cost'])  # Scatter Plot Shows relationship between two variables
plt.plot(final_data['Distance_km'], model.predict(X)) #Draws a line over the scatter plot
plt.show()
```
